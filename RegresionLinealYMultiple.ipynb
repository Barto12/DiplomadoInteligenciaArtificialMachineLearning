üßë‚Äçüè´ C√≥mo explicarlo:
‚ÄúEs como si ense√±aras a una m√°quina a predecir resultados d√°ndole ejemplos ya resueltos. Por ejemplo, si le das muchas casas con su tama√±o y precio, ella aprende la relaci√≥n para predecir el precio de una casa nueva.‚Äù

C√≥digo:
 ejemplo de regresi√≥n lineal simple con datos de casas.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Datos simples: Tama√±o de casa (m2) vs Precio (en miles)
X = np.array([[50], [60], [70], [80], [90], [100], [110]])
y = np.array([150, 170, 190, 210, 230, 250, 270])

# Modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Predicci√≥n
y_pred = modelo.predict(X)

# Visualizaci√≥n
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='L√≠nea de regresi√≥n')
plt.xlabel("Tama√±o de casa (m2)")
plt.ylabel("Precio (mil)")
plt.legend()
plt.show()

# Evaluaci√≥n
print("MSE:", mean_squared_error(y, y_pred))
print("R¬≤:", r2_score(y, y_pred))
```

Regresi√≥n Lineal Simple ‚ÄúSirve para predecir un valor num√©rico a partir de una sola variable. Es como trazar una l√≠nea recta que mejor se ajusta a los puntos de nuestros datos.‚Äù

üß© Analog√≠a:
‚ÄúImaginen que tienen una regla transparente y tratan de colocarla donde mejor pase por los puntos. Eso hace el algoritmo, pero usando matem√°ticas.‚Äù

ejemplo de tama√±o de casa vs precio.

Resalta:

.fit(X, y) = aprende

.predict(X) = predice

Mostrar el gr√°fico y analizar juntos la l√≠nea.

```python
# Paso 1: Importar librer√≠as
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Paso 2: Crear los datos (tama√±o en m¬≤ y precio en miles de pesos)
X = np.array([[50], [60], [70], [80], [90], [100], [110]])  # Tama√±o de casa
y = np.array([150, 170, 190, 210, 230, 250, 270])            # Precio

# Paso 3: Crear el modelo y entrenarlo
modelo = LinearRegression()
modelo.fit(X, y)

# Paso 4: Predecir con el modelo entrenado
y_pred = modelo.predict(X)

# Paso 5: Visualizar los resultados
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='L√≠nea de regresi√≥n')
plt.xlabel("Tama√±o de casa (m¬≤)")
plt.ylabel("Precio (miles de pesos)")
plt.title("Regresi√≥n Lineal Simple")
plt.legend()
plt.grid(True)
plt.show()

# Paso 6: Evaluar el modelo
print("Coeficiente (pendiente):", modelo.coef_[0])
print("Intersecci√≥n (ordenada al origen):", modelo.intercept_)
print("Error Cuadr√°tico Medio (MSE):", mean_squared_error(y, y_pred))
print("Coeficiente de determinaci√≥n (R¬≤):", r2_score(y, y_pred))
```

 Evaluaci√≥n del Modelo ‚ÄúDespu√©s de predecir, necesitamos saber qu√© tan buena fue la predicci√≥n. Para eso usamos m√©tricas.‚Äù

‚úçÔ∏è Explica dos m√©tricas:
MSE (error cuadr√°tico medio): qu√© tan lejos est√°n nuestras predicciones.

R¬≤ (R-cuadrado): qu√© tanto explica nuestro modelo del comportamiento de los datos. (1 = perfecto)

Regresi√≥n Lineal M√∫ltiple ‚ÄúAqu√≠ usamos varias variables para predecir. Es como decir: ‚ÄòEl precio de una casa no solo depende del tama√±o, tambi√©n del n√∫mero de habitaciones y la ubicaci√≥n‚Äô.‚Äù

üß© Analog√≠a:
‚ÄúNo es lo mismo una pizza por su tama√±o, tambi√©n influye si tiene m√°s ingredientes.‚Äù

‚úÖ C√≥digo:
Usa el ejemplo con tama√±o y n√∫mero de habitaciones.

Preguntae a los alumnos: ¬øCu√°l de las dos variables creen que influye m√°s?

```python
# Paso 1: Importar librer√≠as
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Paso 2: Datos
# Columnas: [tama√±o (m¬≤), n√∫mero de habitaciones]
X = np.array([
    [50, 2],
    [60, 2],
    [70, 3],
    [80, 3],
    [90, 4],
    [100, 4]
])

# Precio de las casas (en miles)
y = np.array([160, 180, 200, 220, 240, 260])

# Paso 3: Crear el modelo y entrenarlo
modelo = LinearRegression()
modelo.fit(X, y)

# Paso 4: Hacer predicciones
y_pred = modelo.predict(X)

# Paso 5: Evaluaci√≥n del modelo
print("Coeficientes (pendientes):", modelo.coef_)
print("Intersecci√≥n (ordenada al origen):", modelo.intercept_)
print("MSE:", mean_squared_error(y, y_pred))
print("R¬≤:", r2_score(y, y_pred))

# Paso 6: Probar con un nuevo dato
nueva_casa = np.array([[85, 3]])  # Ej: 85 m¬≤ y 3 habitaciones
prediccion = modelo.predict(nueva_casa)
print("Precio estimado para casa de 85 m¬≤ con 3 habitaciones:", prediccion[0])
```

Regularizaci√≥n (L1 y L2) ‚ÄúCuando un modelo aprende demasiado bien, puede memorizar y fallar con nuevos datos. Regularizar es como ponerle un l√≠mite para que no se pase.‚Äù

üß© Analog√≠a:
L1 (Lasso): ‚ÄúEs como hacer una maleta para viajar, pero solo puedes llevar lo esencial. El modelo decide qu√© dejar fuera.‚Äù

L2 (Ridge): ‚ÄúEs como controlar el volumen de cada variable para que ninguna grite m√°s de la cuenta.‚Äù

‚úÖ C√≥digo:
Compara el MSE de Lasso vs Ridge usando los mismos datos que en la regresi√≥n m√∫ltiple

```python
# Paso 1: Importar librer√≠as
import numpy as np
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error

# Paso 2: Datos de entrada
# X = [tama√±o en m¬≤, n√∫mero de habitaciones]
X = np.array([
    [50, 2],
    [60, 2],
    [70, 3],
    [80, 3],
    [90, 4],
    [100, 4]
])
y = np.array([160, 180, 200, 220, 240, 260])  # Precio (en miles)

# Paso 3: Crear modelos con regularizaci√≥n

# Modelo Lasso (L1)
modelo_lasso = Lasso(alpha=0.1)
modelo_lasso.fit(X, y)
pred_lasso = modelo_lasso.predict(X)

# Modelo Ridge (L2)
modelo_ridge = Ridge(alpha=1.0)
modelo_ridge.fit(X, y)
pred_ridge = modelo_ridge.predict(X)

# Paso 4: Evaluar modelos
print("=== Resultados ===")
print("Lasso Coeficientes:", modelo_lasso.coef_)
print("Lasso MSE:", mean_squared_error(y, pred_lasso))

print("Ridge Coeficientes:", modelo_ridge.coef_)
print("Ridge MSE:", mean_squared_error(y, pred_ridge))
```

Aplicaciones Pr√°cticas (Notas escolares) ‚ÄúVeamos c√≥mo se puede aplicar esto para predecir las calificaciones de un estudiante seg√∫n sus h√°bitos.‚Äù

‚úÖ C√≥digo:
Usa el ejemplo de horas de estudio, sue√±o y asistencia.

```python
# Paso 1: Importar librer√≠as
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Paso 2: Crear los datos (Ejemplos de estudiantes)
# Columnas: [horas de estudio, horas de sue√±o, clases asistidas]
X = np.array([
    [2, 8, 10],
    [3, 7, 9],
    [4, 6, 10],
    [5, 6, 9],
    [6, 5, 8],
    [7, 5, 10],
    [8, 4, 9]
])

# Notas obtenidas por cada estudiante
y = np.array([60, 65, 70, 75, 80, 85, 90])

# Paso 3: Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Paso 4: Predecir valores
y_pred = modelo.predict(X)

# Paso 5: Evaluar el modelo
print("Coeficientes:", modelo.coef_)
print("Intersecci√≥n (bias):", modelo.intercept_)
print("MSE:", mean_squared_error(y, y_pred))
print("R¬≤:", r2_score(y, y_pred))

# Paso 6: Probar con un nuevo alumno
nuevo_estudiante = np.array([[5, 6, 10]])  # 5h estudio, 6h sue√±o, 10 clases asistidas
prediccion = modelo.predict(nuevo_estudiante)
print("Nota estimada para el nuevo estudiante:", prediccion[0])
```
