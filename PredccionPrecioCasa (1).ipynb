```python
# Ejemplo sencillo de regresiÃ³n con scikit-learn
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Datos ficticios: metros cuadrados de casas
X = np.array([[50], [65], [80], [100], [120], [140], [160], [180]])
# Precios correspondientes (en miles de $)
y = np.array([100, 150, 200, 250, 300, 350, 400, 450])

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Predecir para un nuevo tamaÃ±o de casa: 110mÂ²
precio_predicho = modelo.predict([[110]])
print(f"Precio predicho para 110mÂ²: ${precio_predicho[0]:.2f} mil")

# Visualizar la regresiÃ³n
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, modelo.predict(X), color='red', label='LÃ­nea de regresiÃ³n')
plt.scatter([[110]], precio_predicho, color='green', s=100, label='PredicciÃ³n')
plt.xlabel('Metros cuadrados')
plt.ylabel('Precio (miles $)')
plt.title('RegresiÃ³n Lineal: Precio vs TamaÃ±o')
plt.legend()
plt.grid(True)
plt.show()

# Coeficientes del modelo
print(f"Pendiente: {modelo.coef_[0]:.2f}")
print(f"Intercepto: {modelo.intercept_:.2f}")
```
